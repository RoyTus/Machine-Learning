```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x,y = make_moons(700,noise=0.4, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=42)
plt.title('Half moon Data set')
plt.scatter(X_train[:,1],X_train[:,0],c=y_train)
plt.show()
```

```python
logreg=linear_model.LogisticRegression(C=5)
logreg.fit(X_train, y_train)
x1=logreg.coef_
w0=logreg.intercept_
w1=x1[0,0]
w2=x1[0,1]

# plot the line showing 
plt.figure()
plt.title('Linear Classifier used on the Training data Set')
plt.plot([-2,(-(2*w2+w0)/w1)],[((-w0+2*w1)/w2),2])
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, edgecolors='k', cmap=plt.cm.Paired)
plt.show()
```

```python
Z = logreg.predict(X_test)
plt.figure()
plt.title('Linear Classifier used on the Testing Data set and the Output')
plt.plot([-2,(-(2*w2+w0)/w1)],[((-w0+2*w1)/w2),2])
plt.scatter(X_test[:, 0], X_test[:, 1], c=Z, edgecolors='k', cmap=plt.cm.Paired)
plt.show()
#print(Z)
```

```python
print(metrics.accuracy_score(y_test, Z))
```
